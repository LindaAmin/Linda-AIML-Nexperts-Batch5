### Large Language Models (LLMs)

What is an LLM?

Large Language Models:
- Trained on massive text data - 
- Predict next word/token - 
- Use billions of parameters - 
    - `Claude Mythos 5` recently surpassed 10 trillion parameters

**How Large Language Models (LLMs) Work?**

- core concept of text generation: 
    - next-token prediction
- 
- a maths machine that continously and at really fast speed calculates the probability on the next word

- How an AI doesn't "think" in sentences, but rather calculates the **statistical probability** of what the next word (or piece of a word) should be, based on the massive amounts of text it has read.
- 
- Most LLMs work on the concept of **Transformers** 
- Analogy of reading a book: 
    - instead of reading word-by-word 
    - forgetting the start of the page, 
    - the "**attention mechanism**" allows the AI to pay attention to all **relevant words** in a sentence simultaneously to understand context
