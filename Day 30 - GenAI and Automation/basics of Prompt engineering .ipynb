## Prompt Engineering
- 
- 

- A prompt is a query to a LLM
- Large Language Models
    - Language Models are DL models that can deal with Language
        - Understanding Language
        - Generating Language
    - LLMs are trained on massive amounts of data
    - LLMs are the engines behind ChatGPT, Gemini, Claude and Deepseek
    - Think of them as:
        - Really really advanced Auto Complete
        - Rather a search engine. Or a person.
    - Next token generation --- basically predicts the next best word to use based on its training
    - 
- At the core of their functioning, they are just massive statistical models - 
- they calculate probability of the next word in the sequence -> predict the best next word 

### Why the same prompt can give you different results
- simple calulator will return 4 every time you ask 2 + 2
- but in LLMS the response depends on two things: 
    1. TEMPERATURE
    2. RANDOMNESS

1. Next Token Prediction:
- 
- predict the token when you give the prompt --> a probibility distribution is returned
- Imagine the model is completing the sentence: `The cat sat on the...`

    - `Mat`: 80% probability
    - `Floor`: 15% probability
    - `Pizza`: 5% probability
- If the model always chose the 80% option, it would be extremely **boring** and **repetitive**. 
- To make it creative and human-like, we use settings to let it "roll the dice."

2. **Temperature: The "Chaos" Dial**
    - Temperature is a configuration setting that flattens or sharpens that probability distribution.

**Temperature** - The chaos dial
- setting that flattens or sharpens the probability distribution
- 
- Low Temperature: (0.1 to 0.3)
    - The model is greedy
    - always picks the highest probability words
- 
- High Temperature: (0.7 - 1+)
    - The model takes more risks
    - It might pick the 15% or 5% options too.
    - Leads to more diverse, creative, and sometimes **hallucinatory** responses

### importance of CONTEXT

**Context Window**:
- the number of tokens an LLM can keep in memory at an instant
- Active Memory - the is being used all the time
    - in computers one would call RAM as the active memory
    - Working memory in humans - e.g. the amount on info you can keep in your head during a conversation
- If a model has **128K context window** - we can fit apx a 400 page book
- the more we fill this context window, the higher the latency(longer it will take to respond) more the cost too
- 
- Think of the context window as the **digital whiteboard** the AI can see while it’s talking to you.
- Every message you send, and every response the AI gives, is written on that whiteboard.
- The AI "re-reads" the entire whiteboard every single time it generates a new word.

- The Catch: The whiteboard has a physical limit (measured in tokens). 
    - Once it’s full, the oldest information at the top, is "erased" to make room for new notes
    - just like humans forgetting

### **Lost in the Middle Phenomemon**
- Research points out that, LLMs are not equally good at reading the whole context window
- 
- They have a U-shaped attention span:
    - **High Attention**: on the information at the `beginning` (initial instructions)
    - **Low Attention**: on the information buried in the `middle` of a conversation or a 50 page pdf
    - **High Attention**: on the information at the very `end` (The most recent prompt)

- Prompt Engineering Tip: Keep the mission-critical instructions at the very botton of your prompt, right before the AI to execute

### Garbage In - Garbage Out

The quality of the output is directly limited by the quality and clarity of the input.

The biggest cause of "Garbage Out" is a lack of constraints. An AI is a `**yes-man**`; it will try to fulfill any request, even a bad one.

- **Garbage In**: "`Write a blog post about marketing.`"

    - Result: A generic, boring 300-word essay that sounds like a Wikipedia entry from 2012.

- **Quality In**: "`Write a 500-word blog post for B2B SaaS founders about using LinkedIn Ads to lower Customer Acquisition Cost (CAC). Use a punchy, authoritative tone and include a bulleted list of 3 common mistakes.`"
    - Result: A targeted, professional piece of content that actually provides value

- **The Three Pillars of "Clean" Input**

To avoid the GIGO, a prompt need to pass the P.A.S. test:
- **Precision**: Use strong verbs and exact numbers (write 3 paragraphs instead of write a bit)
- **Audience**: Tell the AI who it is writing for. 
- **Structure**: Give the AI a skeleton to frame the outputs ("Use headers, a table of contents, and a concluding summary)

### Anatomy of a Prompt
- refers to the the building blocks of a prompt
- things used to get teh best and most accurate outputs
- 
1. **Role:** 
- Tell the AI who to act like
    - "Act like a senior Python developer..."
2. **Context:** 
- Provide background information or constraints to steer the model
    - "...working on a customer churn prediction project"
3. **Instruction/Directive/Task:**
- The specific action you want to take like Summarize, Write the code, Analyze, Write a rhyming poem
    - "Write a function to clean this dataset"
4. **Input Data:**
- The raw information to be processed
- A specific piece of content that you want processed like text, csv, ...
    - "[Paste a CSV data file, or CSV code snippet]"
5. **Output Format:** 
- The format of the the output result
- How you want the result to be delivered
    - "Return only the code block with comments"
    - 
    1. *Desired File Format*: in bullet points, as a JSON object, a text file, a word document, an image
    2. *Formatting*: Markdown, JSON, XML, CSV
    3. *Length*: Specific word-count, paragraph limits, one-sentence answers
    4. *Tone & Style*: Mimicking a specific author or brand voice, or use a certain style of language. 

**Example of Stuructured Prompts**

**Weak Prompt**

"Tell me how to build a machine learning model."

Result: You'll get a generic, 10-page essay that might not even apply to your specific tools.


**The Structured Prompt:**


`[Role]` Act as a Lead Data Scientist.
`[Context]` I am building a binary classification model using the Telco Customer Churn dataset in a Jupyter Notebook.
`[Task]` Help me perform feature engineering.
`[Instructions]` Focus on encoding categorical variables like 'Gender' and 'Contract' using One-Hot Encoding. Please ensure the code is compatible with Scikit-Learn.
`[Output]` Provide the Python code in a single block with a brief explanation of why we chose One-Hot Encoding over Label Encoding.

> **Tips: **
- *Be Direct*: Use `Do` instead of `Don't`
e.g. `Write a short summary` works better than `Don't write a long summary`
- *Use Few-Shot Prompting*: Provide 1 or 2 examples of the output you want
- *Demlimiters*: help in adding clarity. Use `(""")` , backticks `(```)`, or XML tags (`<data> ... </data>`)

### **2. Core Frameworks**

Framework|Meaning|Best For...
---|---|---
R-T-F|Role, Task, Format|Quick, everyday tasks.
C-A-R-E|Context, Action, Result, Example|Complex business requests.
C-O-S-T|Context, Objective, Style, Tone|Creative writing and branding.

### **3. Advanced Strategy Techniques**


This is where the real "engineering" happens. These techniques help the model "think" more accurately.

**Few-Shot Prompting:**
- Providing Examples
- Providing 2-3 examples of the desired input/output behavior before asking for the result.

**Chain of Thought (CoT):** 
- Step-by-Step Instructions
- Asking the model to "show its work" or "think step-by-step." This significantly reduces logic errors in math or reasoning.
- For complex tasks, ask the model to "think step-by-step" or provide a sequence of instructions.

**Delimiters:** 
- Using characters like """, ---, or < > to clearly separate instructions from the data being analyzed.

**Role Prompting:** 
- Assigning a specific persona 
- "Act as a [`role`, e.g., expert historian, career coach, Socratic tutor]."
- (e.g., "Act as a Senior DevOps Engineer with 20 years of experience").

### **4. Ethics, Constraints & Responsible Use**

- Teaching how to tell the AI what not to do is just as important as telling it what to do.

- Bias: LLMs reflect the biases present in their training data. How to identify and mitigate biased outputs.

- Negative Constraints: 
- In text generation 
- "Do not use jargon," or "Do not mention [`Competitor Name`]."

- Hallucination Checks: 
    -  LLMs can be confidently wrong. The importance of critical thinking and verification.
    - Techniques like "`If you don't know the answer, state that you don't know`" to prevent the AI from making things up

- Privacy & Data Security: Never put sensitive personal or proprietary information into public AI tools.

- Plagiarism & Originality: Using AI as a tool for inspiration and drafting, not for copying. Understanding the model's role as a creative collaborator, not an author.
