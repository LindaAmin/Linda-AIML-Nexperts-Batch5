```python
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori, association_rules
```

```python
# dataset of transaction
transactions = [
    ['Milk','Butter','Bread'],
    ['Milk','Bread'],
    ['Bread','Butter','Eggs'],
    ['Butter','Eggs'],
    ['Bread','Butter','Juice']
]
```

```python
# EDA: Convert the transaction data into a one-hot encoded DataFrame
# Data Cleaning: Convert the transaction data into a one-hot encoded DataFrame
# Label encoding: Convert the transaction data into a one-hot encoded DataFrame 
# Algorithm cannot read words, so we need to convert the transaction data into a one-hot encoded DataFrame. One-hot encoding is a process of converting categorical data into a binary matrix. In this case, we will convert the transaction data into a binary matrix where each row represents a transaction and each column represents an item. If an item is present in a transaction, the corresponding cell will be 1, otherwise it will be 0.

te = TransactionEncoder()

#te_ary = te.fit(transactions).transform(transactions)
te_array = te.fit_transform(transactions)

# create a dataframe of these encoded transaction along with the table header
df_transactions = pd.DataFrame(te_array, columns=te.columns_)
display(df_transactions)
```

```python
# Help frequently bought items in the transaction data using the Apriori algorithm. The Apriori algorithm is a classic algorithm used in data mining for learning association rules. It is used to find frequent itemsets in a dataset and derive association rules from them. The algorithm uses a breadth-first search strategy to count the support of itemsets and prune the search space based on the minimum support threshold.
# run the Apriori algorithm on the one-hot encoded DataFrame to find frequent itemsets with a minimum support of 0.4

frequent_itemsets = apriori (df_transactions, # the one-hot encoded DataFrame
                            min_support=0.4, # minimum support of 0.4 means that the itemset must appear in at least 40% of the transactions
                            use_colnames=True) # use_colnames=True means that the itemsets will be returned as a list of items instead of a list of column indices

```

```python
frequent_itemsets
```

```python
clean_itemsets = frequent_itemsets['itemsets']
```

```python
clean_itemsets
```

```python
# Generate association rules from the frequent itemsets using the association_rules function. The association_rules function takes in the frequent itemsets and generates a set of rules based on the specified metric and threshold. In this case, we are using confidence as the metric and a minimum threshold of 0.7.
# rules to find what is likely going to be bought together. The association rules are generated based on the frequent itemsets found in the previous step. The rules are generated based on the specified metric and threshold. In this case, we are using confidence as the metric and a minimum threshold of 0.7. Confidence is a measure of how often the rule has been found to be true. It is calculated as the support of the itemset divided by the support of the antecedent.
rules = association_rules(frequent_itemsets, # dataframe of itemsets & supportgenerated by the apriori algorithm 
                          metric="confidence", # metric to evaluate the quality of the rules. In this case, we are using confidence as the metric. Confidence is a measure of how often the rule has been found to be true. It is calculated as the support of the itemset divided by the support of the antecedent.
                          min_threshold=0.7) # at least 70% confidence means that the rule must be true at least 70% of the time

final_table = rules[["antecedents", "consequents", "support", "confidence", "lift"]]

display(final_table)
```
