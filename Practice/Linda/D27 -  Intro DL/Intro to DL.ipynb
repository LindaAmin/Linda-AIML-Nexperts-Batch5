```python
import numpy as np
 
inputs =  np.array([1, 1, 1, 0, 0]) # depends on situation # can be anything
weights = np.array([0.7, 0.5, 0.6, 0.3, 0.4 ]) # we assign weights ourselves

# let's apply the frankrosenblat algo

# Step 1: Threshold
threshold = 1.5

# Step 2: Multiply inputs with weights
product = inputs * weights # broadcasting 
product

# Step 3: Weighted sum
weighted_sum = sum(product)

# Step 4: Activation function
def step_activation(x):
    if x >= threshold:
        return "Yes, we will go to the concert"
    else:
        return "No, we are not going"

# apply the activation
step_activation (weighted_sum)
```
